```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Emisiones 2020.xlsx',"VOC",index_col=None)
df2 = pd.read_excel('input/GEIA_VOC.xlsx',index_col=None)
```

```python
df2
```

```python
dfm=df.merge(df2, left_on=["Categoría","Motorizacion","Norma"], right_on=["Modo","Motorizacion","Norma"], how="left")
dfm = dfm.fillna(0)
VOCs = dfm.loc[:,"Ethane":"Alkanones (ketones)"]
#dfm.loc[:,2014:2050].multiply(dfm.loc[:,"Ethane"], axis="index")
s={}
for i in VOCs.columns.values:
    df_f = dfm.loc[:,2014:2050].multiply(dfm.loc[:,i], axis="index")/100
    s[i] = pd.concat([dfm.iloc[:,:7],df_f],axis=1)
```

```python
writer = pd.ExcelWriter('Emisiones 2020 VOC.xlsx')
for i in VOCs.columns.values:
    s[i].to_excel(writer,i)
writer.save()
```
